```python
import pandas as pd
```

```python
df = pd.read_csv('Customer-Churn-Prediction - Customer-Churn-Prediction.csv')
```

```python
df.head()
```

# Uisng Null Method

```python
df.isnull().sum()
```

```python
null_mask = df.isnull().any(axis=1)
null_rows = df[null_mask]

print(null_rows)
```

# Using Query Method 

```python
# We can also use the query() method to select rows based on null values. 
#This method allows us to write SQL-like queries on the dataframe, 
#making it easy to select rows based on certain conditions.

null_rows = df.query('TotalCharges != TotalCharges')

print(null_rows)
```

# Using loc Method

```python
# Finally, we can also use the loc method to select rows based on null values. 
# This method allows us to select rows and columns by label or boolean mask.

null_rows = df.loc[df['TotalCharges'].isnull()]

print(null_rows)
```

# For Not NULL
## Using notnull() Method

```python
df.notnull().sum()
```

```python
"""We can also select rows based on non-null values using the notnull() method. 
This method is the opposite of isnull() and returns a boolean mask indicating whether 
each element in the dataframe is not null."""

not_null_mask = df.notnull().all(axis=1)
not_null_rows = df[not_null_mask]

print(not_null_rows)
```
